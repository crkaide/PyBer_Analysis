```python
# Import dependency
# Chain pyplot module to it
# essentially saying "from matplotlib import pyplot"
# assign alias plt

import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.

x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.

y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot

plt.plot(x_axis, y_axis)
```

```python
# When we switch from using MATLAB to the object-oriented interface method, we use ax.plot()
# instead of plt.plot()
```

```python
# OBJECT-ORIENTED interface method
```

```python
# Create the plot with ax.plt()

fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)

# Can also be written as:
# fig = plt.figure()
# ax = fig.add_subplot()
```

```python
# Create the plot with ax.plt()

# CREATES THE EMPTY GRAPH
fig = plt.figure()
ax = fig.add_subplot()

# PLOTS THE INFORMATION CONTAINED WITHIN THE GRAPH
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)

# To change figure-level attributes (such as axis labels, a title, or a legend)
# or save the figure as an image, use fig = plt.figure()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()

# plt.show() should only be used once per session because it can cause the
# graph to display improperly. If you have problems with data not being
# displayed properly, you may have to restart your kernel and clear your output.
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45.
plt.ylim(0, 45)

# Create a title.
plt.title("PyBer Fare by Month")

# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45.
plt.ylim(0, 45)

# Create a title.
plt.title("PyBer Fare by Month")

# Add a grid.
plt.grid()

# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.barh(x_axis, y_axis)

# To invert the y-axis to have the months in ascending order, use the gca() method.
# The gca() method means "get current axes." We can chain the gca() method to the
#invert_yaxis() method by using gca().invert_yaxis()
plt.gca().invert_yaxis()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Scatter chart plt.plot() method

plt.plot(x_axis, y_axis, 'o')
```

```python
plt.scatter(y_axis, x_axis, color = 'red')
plt.legend(["Chicago"])
plt.title("PyBer Fare by Month")
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.xlim(0, 50)
plt.gca().invert_yaxis()
```

```python
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
    
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# List comprehension function

plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s=[i * 5 for i in y_axis], color = 'skyblue', alpha = 0.2, edgecolor = 'black', linewidth = 2)
ax.legend(["Boston"])
ax.set_title("PyBer Fare by Month")
ax.set_xlim(0, 50)
plt.gca().invert_yaxis()
```

```python
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
colors = ['orange','darksalmon','aliceblue','thistle','indigo','royalblue','olive','blue','rebeccapurple','plum','chocolate','tomato']
explode_values_new = [0,0,0.2,0,0,0,0.2,0,0,0,0,0]
fig, ax = plt.subplots(figsize=(8, 8))
ax.pie(y_axis,labels=x_axis,explode=explode_values_new,autopct='%.1f%%',shadow=True,startangle=85, colors=colors)
plt.gca().invert_xaxis()
plt.show()
```
